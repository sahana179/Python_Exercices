```python
#Generate the following array without explicit loops and without manually fill in the numbers. We will name it arr1:
import numpy as np

arr1 = np.arange(25).reshape(5,-1).T**2
print(arr1)

#Extract the diagonal values of arr1 (do not generate it) and call it diag1
diag1 = arr1[np.arange(5), np.arange(5)]
print(diag1)

#Alternatively
arr = np.arange(0, 25)
arr1 = arr.reshape((5, 5), order='F')
arrayf = arr1*arr1

print(arrayf)
arrayf.diagonal()
```

```python
#extract the submatrix and name it subarr1:
subarr1 = arr1[[0, 2, 1], :][:, [3, 2, 1]]
print(subarr1)
```

```python
#Generate the following array without explicit loops and without manually fill in the numbers. We will name it arr2
arr = np.arange(5, 0, -1)
arr2 = np.tile(arr, (5, 1)).T + arr
print(arr2)

#Replace all the even values of arr2 by 0 without explicit loops.
arr2[arr2%2==0] = 0
print(arr2)
```

```python
#Generate and (scatter) plot 2000 random points in the space [−10,0]×[1,2] with the color orange
import numpy as np
import matplotlib.pyplot as plt

n = 2000
x = -10 * np.random.rand(n)
y = np.random.rand(n) + 1

plt.scatter(x, y, color="orange");
```

```python
#color the points inside the space [−4,−2]×[1.2,1.8] or [−8,−6]×[1.2,1.8] by green.
import numpy as np
import matplotlib.pyplot as plt

ind1 = (x<=-2) & (x>=-4) & (y<=1.8) & (y>=1.2)
ind2 = (x<=-6) & (x>=-8) & (y<=1.8) & (y>=1.2)
x_green = x[ind1 | ind2]
y_green = y[ind1 | ind2]

plt.scatter(x, y, color="orange");
plt.scatter(x_green, y_green, color="green");
```
